```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        print(self.name + " says Woof!")


my_dog = Dog("Buddy", 3)
another_dog = Dog("Max", 5) 
another_dog.bark()
my_dog.bark()

        
```
