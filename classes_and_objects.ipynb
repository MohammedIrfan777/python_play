```python
class Dog:
    legs = 4  # Class variable shared by all instances
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.legs = 4

    def bark(self):
        return f"{self.name} says Woof!"
    

myDog = Dog("Buddy", 5)
print(myDog.name)  # Output: Buddy  
print(myDog.age)   # Output: 5
print(myDog.legs)  # Output: 4

Dog.legs
```

```python
class Dog:
    _legs = 4  # Class variable shared by all instances
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self._legs = 5

    def getLegs(self):
        return self._legs

    def bark(self):
        return f"{self.name} says Woof!"
    

myDog = Dog("Buddy", 5)


print(myDog._legs)
print(myDog.getLegs())
print(Dog._legs)
```

```python
class WordSet:
    def __init__(self, words):
        self.words = set(words)

    def addText(self, text):
        text = WordSet.cleanText(text)
        for word in text.split():
            self.words.add(word)

    @staticmethod
    def cleanText(text):
        return text.lower().replace('.', '').replace(',', '')
    

myWord = WordSet([])
myWord.addText("Hello, world. Hello!")
print(myWord.words)
```
