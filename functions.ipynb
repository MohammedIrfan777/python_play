```python
def addition(a, b, operation='add'):
    if operation == 'add':
        return a + b
    elif operation == 'subtract':
        return a - b
    elif operation == 'multiply':
        return a * b
    elif operation == 'divide':
        return a / b
    else:
        raise ValueError('Invalid operation')   
    
addition(10, 5, 'add')
# addition(10, 5, 'subtract') 
# addition(10, 5, 'multiply')
# addition(10, 5, 'divide')
```
