```python
# Python codeвЂ‹вЂ‹вЂ‹вЂ‹вЂ‹вЂ‹вЂЊвЂЊвЂЊвЂЊвЂ‹вЂЊвЂ‹вЂЊвЂ‹вЂ‹вЂЊвЂ‹вЂЊвЂЊвЂ‹вЂ‹вЂ‹вЂЊвЂЊвЂ‹вЂ‹вЂ‹вЂЊвЂЊвЂЊ below



def encodeString(stringVal):

    if not stringVal:
        return[]

    result = []
    prev = stringVal[0]
    count = 1

    for i in range(1, len(stringVal)):
        if stringVal[i] == prev:
            count +=1
        else:
            result.append((prev, count))
            prev = stringVal[i]
            count = 1

    result.append((prev, count))
    return result


def decodeString(encodedList):
    
    result = ""

    for char, count in encodedList:
        result += char * count

    print(result)
```

```python
encodeString('aaabbcaaa')
```

```python
decodeString([('a', 3), ('b', 2), ('c', 1), ('a', 3)])
```
